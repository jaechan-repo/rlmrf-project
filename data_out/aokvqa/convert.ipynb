```python
import json

INPUT_FILE = "aokvqa_pred_baseline.json"
OUTPUT_FILE = "conv_" + INPUT_FILE

with open(INPUT_FILE, 'r') as f:
    data = json.load(f)

data_converted = {}
for line in data:
    data_converted.update(line)

with open(OUTPUT_FILE, 'w') as f:
    json.dump(data_converted, f)
```

```python
import json
paths = [
    "/mmfs1/gscratch/ark/chan0369/rampa-project/data_out/aokvqa/baseline/predictions_test.json",
    "/mmfs1/gscratch/ark/chan0369/rampa-project/data_out/aokvqa/reward_text/predictions_test.json"
]
test_data_path = "/mmfs1/gscratch/ark/chan0369/rampa-project/data_in/iter0/aokvqa/test.json"
with open(test_data_path, 'r') as f:
    test_data = json.load(f)


submissions = []
for path in paths:
    with open(path, 'r') as f:
        old_submission = json.load(f)
    submission = {}
    for line in test_data:
        qid = line['question']['qid']
        choices = line['question']['choices']
        correct_idx = old_submission[qid]['multiple_choice']
        correct_answer: str = choices[correct_idx]
        submission[qid] = {
            "multiple_choice": correct_answer,
            "direct_answer": ""
        }
    submissions.append(submission)
```

```python
for path, submission in zip(paths, submissions):
    with open(path, 'w') as f:
        json.dump(submission, f)
```
